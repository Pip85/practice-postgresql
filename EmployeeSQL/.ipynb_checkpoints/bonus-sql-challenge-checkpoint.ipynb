```python
# Import postgres password and pandas module

from config import password
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create engine to import sql_challenge database from postgresql

from sqlalchemy import create_engine
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/sql_challenge')
connection = engine.connect()

# Import tables from sql_challenge database in postgresql
salaries_df = pd.read_sql('SELECT * FROM salaries', engine)
titles_df = pd.read_sql('SELECT * FROM titles', engine)
employees_df = pd.read_sql('SELECT * FROM employees', engine)
```

```python
# Calculate max and min of salary range
salaries_df.salary.max()
salaries_df.salary.min()

# Establish salary bins for the histogram
salary_bins = [40000, 50000, 60000, 70000, 80000, 90000, 100000, 130000]

# Create histogram
plt.hist(x=salaries_df['salary'])
plt.title("Salary Ranges by Number of Employees")
plt.xlabel("Salaries")
plt.ylabel("Count of Employees")
plt.show()
```

```python
# Merge employee and salary data
emp_sal_df = pd.merge(employees_df, salaries_df, on="emp_no", how="inner")

# Merge employee and salary data with title data
emp_sal_title_df = pd.merge(emp_sal_df, titles_df, left_on="emp_title_id", right_on="title_id", how="inner")
```

```python
# Calculate average salary by title
avg_sal_title = emp_sal_title_df.groupby("title").mean()["salary"]

# Plot average salary by title
avg_sal_title.plot(kind="bar", rot=45)
plt.title("Average Salary by Title")
plt.xlabel("Job Title")
plt.ylabel("Average Salary")
```
